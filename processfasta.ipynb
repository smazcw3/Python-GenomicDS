```python
#Reading a .fasta file
filename = "data/dna.example.fasta"
```

```python
#Examining the contents of file
!cat data/dna.example.fasta
```

```python
try:
    f = open(filename)
except IOError:
    print("File %s does not exist!!" % filename)

seqs = {}
for line in f:
    line = line.rstrip()  #let's discard the newline at the end (if any)
    if line[0] == '>':  #Distinguising header from sequence
        words = line.split()
        name = words[0][1:]
        seqs[name] = ''
    else: #sequence, not header
        seqs[name] = seqs[name] + line

f.close()
```

```python
# Retrieving data from dictionaries
for name, seq in seqs.items():
    print(name, seq)
```
